#KAGGLE Titanic dataset using Keras Sequential Model

```python
import pandas as pd
import numpy as np
traindata = pd.read_csv("train.csv")
testdata = pd.read_csv("test.csv")

traindata.head()
```

#Drop Names, as they likely do not add valuable information 

```python
X_train = traindata.drop("Name",axis=1).drop("Ticket",axis=1)
X_train['Cabin'] = np.where(X_train['Cabin'].isnull(), 0, 1)
X_train.head(842)
```

```python
X_train = pd.get_dummies(X_train,columns=['Sex','Embarked'])
X_train.head()
```

```python
from sklearn.impute import SimpleImputer
imputer =SimpleImputer(strategy='mean')
#NAN Eintraege durch Mittelwerte ersetzen
X_train['Age']=imputer.fit_transform(X_train['Age'].values.reshape(-1,1))
X_train['Fare']=imputer.fit_transform(X_train['Fare'].values.reshape(-1,1))

X_train.head()
```
